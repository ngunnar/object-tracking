## Introduction

Visual Tracking = Process to detecting the position of an object in each frame of video.

Minimum Output Sum of Squared Error (MOSSE) = a form of Optimized Correlation Output Filter.

There are two main components of the algoritm:

1) Initialization

2) Tracking

Visual tracking is difficult due to variation in the appearance of an object and its surroundings in a video. Object appearance changes due to pose, lighting, natural variations or non ridid transformation. Object may be occluded or move out of the frame of the video.

### Definition of terms

* **Object/Target**
    
    An object is thing which has definable characteristics. An object can also referred as a target


* **Template**

    A sub-image cropped form a larger image. The template is used as an input to initialize the filter. An image of an object is present in the template.


* **Correlation**

    A measure of similarity between two images. The template is moved to different locations in the image and the posistion of the best match is the one with highest correlation.


* **Filter**

    A sub-image placed over the image at different locations and computing pairwise-sum of the filter and corresponding image pixels. When the filter is applied across an entire image the result is itself a new image, i.e. a filtered image.


* **Tracking window**

    A sub-image in which the tracker looks for the new lokcation of the object. 


* **Synthetic target**

    A synthetically genereted image with a Gaussian peak at the location if the object to be tracked. Used to map the input image to its corresponding output to generate a filter.


* **Occlusion**

    Caused by an object which blocks the view of another object in th video.


* **Tracking**

    An action where the algotithm finds the new location of an object in all the frames of a video over time.


* **Initialization**

    A process where the filter used for tracking an object in a video is generated. Can also be referred to as training.


* **Updating**

    A process where a filter is updated with the new information about the object, for example chage in the pose of a person or change in the scale of an object is a new information.


## Optimized Correlation Output Filters

Two Optimized Correlation Output filters (OCOF)

1) Average of Synthetiv Exact filter (ASEF)

2) Minimizing the Output Sum of Squared Error (MOSSE)
